```python
import plotly
plotly.tools.set_credentials_file(username='wwaniek5', api_key='j5UjXoQ2AyVq0DeqKE7u')

import pandas as pd
import sqlite3
import pprint

conn = sqlite3.connect("MouseHeatmapDb.sqlite")
df = pd.read_sql_query("""select 
X,
Y,
cast(MouseFinishedCount as float) as MouseFinishedCount,
cast(MousePassedCount as float) as MousePassedCount,
cast(SpeedCount as float) as SpeedCount 
from ScreenUnits;""", conn)


conn.close()
df=df.groupby(['X','Y'], as_index=False).sum()
df
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np


mouseFinishedCountMatrix= pd.pivot_table(df,values='MouseFinishedCount', index=['Y'],columns=['X'],fill_value =0).values 
mouseFinishedCountMatrix
py.iplot([go.Heatmap(z=np.flipud(mouseFinishedCountMatrix))], filename='mouseFinishedCountHeatmap')
```

```python
MousePassedCountMatrix=pd.pivot_table(df,values='MousePassedCount', index=['Y'],columns=['X'],fill_value =0).values 
py.iplot([go.Heatmap(z=np.flipud(MousePassedCountMatrix))], filename='MousePassedCountHeatmap')
```

```python
py.iplot([go.Heatmap(z=np.flipud(MousePassedCountMatrix-mouseFinishedCountMatrix))], filename='DifferenceHeatmap')
```

```python
SpeedCountMatrix=pd.pivot_table(df,values='SpeedCount', index=['Y'],columns=['X'],fill_value =0).values 
speedMatrix=np.divide(SpeedCountMatrix,MousePassedCountMatrix)
py.iplot([go.Heatmap(z=np.flipud(speedMatrix))], filename='SpeedHeatmap')
```
