```python
import pandas as pd
import sqlite3
import pprint

conn = sqlite3.connect("MouseHeatmapDb.sqlite")
df = pd.read_sql_query("""select X,Y, cast(MouseFinishedCount as float) as MouseFinishedCount from ScreenUnits;""", conn)


conn.close()
df=df.groupby(['X','Y'], as_index=False).sum()

print(type(df.iat[0,2]))
df.sort(columns='MouseFinishedCount', ascending=False)
df.loc[df['X'] == 0]
```

```python
screenMatrix=df.pivot('X', 'Y', 'MouseFinishedCount').values
screenMatrix
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np

trace = go.Heatmap(z=np.rot90(screenMatrix))
data=[trace]
py.iplot(data, filename='basic-heatmap')
```
